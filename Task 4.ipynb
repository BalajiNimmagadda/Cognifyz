# Import necessary libraries

```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
import matplotlib.pyplot as plt
```

# Load your dataset into a Pandas DataFrame (replace 'your_dataset.csv' with your actual file path)


```python
df = pd.read_csv('Dataset.csv')
df
```

# Data Preparation: Clean the dataset (e.g., handle missing data)

```python
df.dropna(subset=['Latitude', 'Longitude'], inplace=True)
df
```

# Create a map centered around a specific location (e.g., the average of all restaurant coordinates)

```python
map_center = [df['Latitude'].mean(), df['Longitude'].mean()]
restaurant_map = folium.Map(location=map_center, zoom_start=10)
```

# Create a MarkerCluster to group restaurant markers for better visualization

```python

marker_cluster = MarkerCluster().add_to(restaurant_map)
```

# Add restaurant markers to the map

```python

for index, row in df.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']], tooltip=row['Restaurant Name']).add_to(marker_cluster)
```

# Save the map to an HTML file

```python

restaurant_map.save('restaurant_map.html')
```

# Grouping by City

```python


city_grouped = df.groupby('City')
```

# Calculate the number of restaurants in each city

```python

city_restaurant_counts = city_grouped.size()
```

# Visualization of restaurant counts by city

```python

city_restaurant_counts.plot(kind='bar', title='Number of Restaurants by City')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.figure(figsize=(50,50))
plt.show()
```
